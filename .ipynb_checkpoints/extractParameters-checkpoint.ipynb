```python
# Make the notebook wider 
from IPython.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline


from Data.dataHelper import load_and_prepare_data


#######################################################################################################
# LOAD AND PREPARE DATASET FOR MODELING
#######################################################################################################

def get_dataset(filepath="Data/dataset.xlsx", features=["P_area", "P_type", "Die_area"]):

    # ------------------------------------------------------------------
    # Load and clean data
    # ------------------------------------------------------------------
    df_train, df_test = load_and_prepare_data(filepath)
    df_test           = df_test.dropna()  # remove one row full of NaN values
    
    df_train["P_type"] = df_train["P_type"].replace("DIP", "SOP/SOT")

    # ------------------------------------------------------------------
    # Keep only the selected features and drop NaNs
    # ------------------------------------------------------------------
    df_train = df_train[features].dropna(subset=features)
    df_test  = df_test[features].dropna(subset=features)

    return df_train, df_test


#######################################################################################################
# CREATE PREDICTION FUNCTIONS FOR THE LINEAR MODEL
# ----------------------------------------------------------------------------------------------------
#    Model: log10(Die_area) = β0 + β1 * log10(P_area) + one-hot(P_type)
#
#    makeSinglePrediction(x, y):
#        Predicts Die_area for a single sample using the fixed numerical 
#        parameters from the fitted linear model (no .fit() or sklearn).
#
#    makeBatchPrediction(df, model_params):
#        Applies makeSinglePrediction() over all rows of a dataset.
#######################################################################################################

def extract_linear_model_parameters(df_train):
    """
    Fit Model 1 on df_train and extract its numerical coefficients.
    """

    # ------------------------------------------------------------------
    # log10(P_area)
    # ------------------------------------------------------------------
    df_train = df_train.copy()
    df_train["log_P_area"] = np.log10(df_train["P_area"])

    # ------------------------------------------------------------------
    # Define input (X) and target (y)
    # ------------------------------------------------------------------
    X_train1 = df_train[["log_P_area", "P_type"]]
    y_train1 = np.log10(df_train["Die_area"])

    # ------------------------------------------------------------------
    # Build preprocessing and model pipeline
    # ------------------------------------------------------------------
    preprocessor = ColumnTransformer([
        ("pass_log_area", "passthrough", ["log_P_area"]),
        ("onehot_type", OneHotEncoder(drop="first", sparse_output=False, handle_unknown="ignore"), ["P_type"])
    ])

    model = Pipeline([
        ("preprocessor", preprocessor),
        ("regressor", LinearRegression())
    ])

    # ------------------------------------------------------------------
    # Fit model
    # ------------------------------------------------------------------
    model.fit(X_train1, y_train1)

    # ------------------------------------------------------------------
    # Extract numerical parameters
    # ------------------------------------------------------------------
    regressor = model.named_steps["regressor"]
    encoder   = model.named_steps["preprocessor"].named_transformers_["onehot_type"]

    categories = encoder.categories_[0][1:]  # because drop='first'
    coefs      = regressor.coef_
    intercept  = regressor.intercept_

    beta_logParea = coefs[0]
    beta_types    = dict(zip(categories, coefs[1:]))

    params = {
        "intercept": intercept,
        "beta_logParea": beta_logParea,
        "beta_types": beta_types,
        "reference_type": encoder.categories_[0][0]  # dropped one
    }

    return params


#######################################################################################################
# EXTRACT MODEL PARAMETERS FOR UNIVARIATE MODEL
# ----------------------------------------------------------------------------------------------------
#    Model: log10(Die_area) = β0 + β1 * log10(P_area)
#
#    extract_univariate_model_parameters(df_train, df_test):
#        Fits a simple linear regression in log–log space (no categorical terms)
#        and returns the numerical coefficients β0 and β1.
#
#    Returned parameters are directly usable by makeSinglePrediction() 
#    or any external function without requiring sklearn.
#######################################################################################################

def extract_univariate_model_parameters(df_train):
    """
    Fit univariate Model 2 on df_train and extract its numerical coefficients.
    """

    # ------------------------------------------------------------------
    # log10(P_area)
    # ------------------------------------------------------------------
    df_train = df_train.copy()
    df_train["log_P_area"] = np.log10(df_train["P_area"])

    # ------------------------------------------------------------------
    # Define input (X) and target (y)
    # ------------------------------------------------------------------
    X_train = df_train[["log_P_area"]]
    y_train = np.log10(df_train["Die_area"])

    # ------------------------------------------------------------------
    # Fit model
    # ------------------------------------------------------------------
    model = LinearRegression()
    model.fit(X_train, y_train)

    # ------------------------------------------------------------------
    # Extract numerical parameters
    # ------------------------------------------------------------------
    intercept = model.intercept_
    beta_logParea = model.coef_[0]

    params = {
        "intercept": intercept,
        "beta_logParea": beta_logParea
    }

    return params

#######################################################################################################
# EXTRACT GROUP-WISE LINEAR MODEL PARAMETERS
# ----------------------------------------------------------------------------------------------------
#    Model (per package type):
#        log10(Die_area) = β0(type) + β1(type) * log10(P_area)
#
#    extract_typewise_model_parameters(df_train):
#        Fits one linear regression per P_type in log–log space
#        and returns intercept and slope for each package type.
#
#    Returned dictionary structure:
#        {
#            "QFP": {"intercept": ..., "slope": ..., "n_samples": ...},
#            "BGA": {"intercept": ..., "slope": ..., "n_samples": ...},
#            ...
#        }
#######################################################################################################

def extract_typewise_model_parameters(df_train):
    """
    Fit a separate linear regression model for each package type (P_type)
    in log–log space and extract numerical coefficients.
    """

    # ------------------------------------------------------------------
    # Clean and prepare data
    # ------------------------------------------------------------------
    df_train = df_train.dropna(subset=["P_area", "Die_area", "P_type"]).copy()
    df_train["log_P_area"]  = np.log10(df_train["P_area"])
    df_train["log_Die_area"] = np.log10(df_train["Die_area"])

    # ------------------------------------------------------------------
    # Fit one regression per package type
    # ------------------------------------------------------------------
    model_params = {}
    for ptype, group in df_train.groupby("P_type"):
        if len(group) < 2:
            continue  # skip small groups

        X = group[["log_P_area"]].values
        y = group["log_Die_area"].values

        reg = LinearRegression()
        reg.fit(X, y)

        model_params[ptype] = {
            "intercept": float(reg.intercept_),
            "slope": float(reg.coef_[0]),
            "n_samples": len(group)
        }

    return model_params
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")

print("\n UNIVARIATE MODEL WITH P AREA")
df_train, df_test = get_dataset("Data/dataset.xlsx", features=["P_area", "Die_area"])
params_univariate = extract_univariate_model_parameters(df_train)
print(params_univariate)

print("\n LINEAR WITH P AREA + P TYPE")
df_train, df_test   = get_dataset("Data/dataset.xlsx", features=["P_area","P_type", "Die_area"])
params_bivariate    = extract_linear_model_parameters(df_train)
print(params_bivariate)

print("\n GROUP-WSIE LINEAR WITH P AREA + P TYPE")
df_train, df_test = get_dataset("Data/dataset.xlsx", features=["P_area","P_type", "Die_area"])
params_group      = extract_typewise_model_parameters(df_train)
print(params_group)
```
