```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")
from Utils.loadData import load_data
import numpy as np
import matplotlib.pyplot as plt


# =============================================================================
# LOAD DATA
# =============================================================================
df_train, df_test = load_data("Data/dataset.xlsx")

df_train = df_train.dropna(subset=["Die_area"])
df_test  = df_test.dropna(subset=["Die_area"])

package_area  = df_test["P_area"].values
die_area_true = df_test["Die_area"].values
P_type_all    = df_test["P_type"].values

x_min = df_test["P_area"].min()
x_max = df_test["P_area"].max()
P_area_line = np.linspace(x_min, x_max, 500)
```

```python
# =====================================================================================
# HELPERS
# =====================================================================================
def _pick_first_existing_column(df, candidates, required=True):
    for c in candidates:
        if c in df.columns:
            return c
    if required:
        raise KeyError(
            "None of the candidate columns were found: "
            + ", ".join([repr(c) for c in candidates])
            + "\nAvailable columns: "
            + ", ".join([repr(c) for c in df.columns])
        )
    return None


def _fit_alpha_ls(x, y):
    """
    Least-squares fit for y ≈ alpha * x (no intercept):
        alpha_LS = sum_i x_i * y_i / sum_i x_i^2
    """
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)
    valid = np.isfinite(x) & np.isfinite(y)
    x = x[valid]
    y = y[valid]
    denom = np.sum(x ** 2)
    if denom <= 0:
        return np.nan
    return np.sum(x * y) / denom


def _fit_alpha_avg_ratio(x, y):
    """
    Average ratio fit:
        alpha_avg = mean(y / x)
    """
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)
    valid = np.isfinite(x) & np.isfinite(y) & (x != 0)
    x = x[valid]
    y = y[valid]
    if x.size == 0:
        return np.nan
    return np.mean(y / x)


# =====================================================================================
# COLUMNS (robust to naming differences)
# =====================================================================================
col_die   = _pick_first_existing_column(df_test, ["Die_area", "die_area", "DIE_area"])
col_ptyp  = _pick_first_existing_column(df_test, ["P_type", "Package_type", "ptype"])
col_parea = _pick_first_existing_column(df_test, ["P_area", "Package_area", "package_area"])

col_mass = _pick_first_existing_column(
    df_test,
    ["Mass", "M"],
    required=False
)

die_area_true = df_test[col_die].values
P_type_all    = df_test[col_ptyp].values


# =====================================================================================
# FIT GLOBAL ALPHAS ON OUR DATA (NON-LOG)
# =====================================================================================
x_area_train = df_train[col_parea].values
y_area_train = df_train[col_die].values

alpha_area_ls  = _fit_alpha_ls(x_area_train, y_area_train)
alpha_area_avg = _fit_alpha_avg_ratio(x_area_train, y_area_train)

print(f"Our data alpha (LS, area):  {alpha_area_ls:.6g}")
print(f"Our data alpha (AVG, area): {alpha_area_avg:.6g}")

alpha_mass_ls = None
alpha_mass_avg = None
if col_mass is not None:
    x_mass_train = df_train[col_mass].values
    alpha_mass_ls  = _fit_alpha_ls(x_mass_train, y_area_train)
    alpha_mass_avg = _fit_alpha_avg_ratio(x_mass_train, y_area_train)
    print(f"Our data alpha (LS, mass):  {alpha_mass_ls:.6g}")
    print(f"Our data alpha (AVG, mass): {alpha_mass_avg:.6g}")


# =====================================================================================
# FIT GROUP-WISE AVG ALPHAS (PROSKE-LIKE, TRAINED ON OUR DATA)
# =====================================================================================
group_order = ["BGA", "QFN/DFN", "QFP", "SOP/SOT", "WLP"]

alpha_area_avg_by_group = {}

for g in group_order:
    mask = (df_train[col_ptyp].values == g)
    x_g = df_train.loc[mask, col_parea].values
    y_g = df_train.loc[mask, col_die].values

    alpha_g = _fit_alpha_avg_ratio(x_g, y_g)
    alpha_area_avg_by_group[g] = alpha_g

    print(
        f"Our data group-wise alpha (AVG, area) [{g}]: "
        f"{alpha_g:.6g}"
    )
```

```python

# =====================================================================================
# MODELS FROM THE LITERATURE + OUR LINEAR FITS
# =====================================================================================
literature_models = {

    # -------------------------------------------------------------------------
    # Teehan (area)
    # -------------------------------------------------------------------------
    "Teehan 2011 (area)": {
        "type": "constant",
        "r": 0.078,
        "input_col": col_parea,
        "input_label": "Package area [mm²]"
    },

    # -------------------------------------------------------------------------
    # Teehan (mass)
    # -------------------------------------------------------------------------
    **({
        "Teehan 2011 (mass)": {
            "type": "constant",
            "r": 18.0,
            "input_col": col_mass,
            "input_label": "Mass [g]"
        }
    } if col_mass is not None else {}),

    # -------------------------------------------------------------------------
    # Proske (literature)
    # -------------------------------------------------------------------------
    "Proske model": {
        "type": "groupwise_constant",
        "input_col": col_parea,
        "input_label": "Package area [mm²]",
        "groups": {
            "BGA":      {"r": 0.16},
            "QFN/DFN":  {"r": 0.28},
            "QFP":      {"r": 0.12},
            "SOP/SOT":  {"r": 0.13},
            "WLP":      {"r": 0.68}
        }
    },

    # -------------------------------------------------------------------------
    # Our data – global linear fits
    # -------------------------------------------------------------------------
    "Our data alpha (LS, area)": {
        "type": "constant",
        "r": float(alpha_area_ls),
        "input_col": col_parea,
        "input_label": "Package area [mm²]"
    },

    "Our data alpha (AVG, area)": {
        "type": "constant",
        "r": float(alpha_area_avg),
        "input_col": col_parea,
        "input_label": "Package area [mm²]"
    },

    # -------------------------------------------------------------------------
    # Our data – group-wise AVG (Proske-like)
    # -------------------------------------------------------------------------
    "Our data group-wise (AVG, area)": {
        "type": "groupwise_constant",
        "input_col": col_parea,
        "input_label": "Package area [mm²]",
        "groups": {
            g: {"r": float(alpha_area_avg_by_group[g])}
            for g in alpha_area_avg_by_group
            if np.isfinite(alpha_area_avg_by_group[g])
        }
    }

}


# =====================================================================================
# MODELS TRAINED ON OUR DATASET (LOG–LOG)
# =====================================================================================
our_models = {

    "Global linear": {
        "type": "global_log",
        "a": -0.46,
        "b":  0.84
    },

    "Linear + category offsets": {
        "type": "offset_log",
        "a": -0.45,
        "b":  0.85,
        "offsets": {
            "BGA":      0.16,
            "QFN/DFN": -0.034,
            "QFP":     -0.18,
            "SOP/SOT": -0.38,
            "WLP":      0.43
        }
    },

    "Group-wise linear (ours)": {
        "type": "groupwise_log",
        "groups": {
            "BGA":     {"a": -0.085, "b": 0.73},
            "QFN/DFN": {"a": -0.67,  "b": 0.99},
            "QFP":     {"a": -0.14,  "b": 0.60},
            "SOP/SOT": {"a": -1.20,  "b": 1.10},
            "WLP":     {"a": -0.12,  "b": 0.99}
        }
    }
}
```

```python
# =====================================================================================
# PLOTTING  (CLEAN REWRITE – ALL MODELS, FIXED ORDER, ALPHA IN BLUE)
# =====================================================================================
def plot_groupwise_models(our_models,
                          literature_models,
                          df_test,
                          die_area_true,
                          P_type_all,
                          loglog=False):

    import numpy as np
    import matplotlib.pyplot as plt

    TITLE_SIZE = 18
    LABEL_SIZE = 16

    group_order = ["BGA", "QFN/DFN", "QFP", "SOP/SOT", "WLP"]
    group_markers = {
        "BGA": "o",
        "QFN/DFN": "s",
        "QFP": "^",
        "SOP/SOT": "D",
        "WLP": "P"
    }

    # ---------------------------------------------------------------------
    # FIXED ORDER – ONLY MODELS THAT EXIST
    # ---------------------------------------------------------------------
    ordered = [
        ("lit", "Teehan 2011 (mass)"),
        ("lit", "Teehan 2011 (area)"),
        ("lit", "Our data alpha (LS, area)"),
        ("lit", "Our data alpha (AVG, area)"),
        ("lit", "Proske model"),
        ("lit", "Our data group-wise (AVG, area)"),  
        ("our", "Global linear"),
        ("our", "Linear + category offsets"),
        ("our", "Group-wise linear (ours)")
    ]

    models = []
    for src, key in ordered:
        if src == "lit" and key in literature_models:
            models.append((key, literature_models[key], "lit"))
        if src == "our" and key in our_models:
            models.append((key, our_models[key], "our"))

    n_cols = len(models)

    fig, axes = plt.subplots(
        3, n_cols,
        figsize=(5 * n_cols, 10),
        gridspec_kw={"height_ratios": [2.5, 1.0, 1.0]},
        sharex=False
    )

    all_abs_err = []
    all_log_err = []

    # =====================================================================
    # LOOP OVER MODELS
    # =====================================================================
    for j, (name, cfg, src) in enumerate(models):

        ax_pred, ax_abs, ax_log = axes[:, j]
        abs_err_all = []
        log_err_all = []

        # =================================================================
        # LITERATURE MODELS (LINEAR)
        # =================================================================
        if src == "lit":

            x_all = df_test[cfg["input_col"]].values.astype(float)
            x_line = np.linspace(np.nanmin(x_all), np.nanmax(x_all), 500)

            is_alpha = name.startswith("Our data")
            line_color = "C0" if is_alpha else "k"
            edge_color = "C0" if is_alpha else "black"
            scatter_color = "C0" if is_alpha else "black"

            if cfg["type"] == "constant":

                r = cfg["r"]
                y_pred_all = r * x_all
                ax_pred.plot(x_line, r * x_line, "--", color=line_color)

                for g in group_order:
                    mask = (P_type_all == g)
                    ax_pred.scatter(
                        x_all[mask], die_area_true[mask],
                        marker=group_markers[g],
                        facecolors="none",
                        edgecolors=edge_color
                    )

                valid = (die_area_true > 0) & (y_pred_all > 0)
                if valid.any():
                    abs_err = y_pred_all[valid] - die_area_true[valid]
                    log_err = np.log10(y_pred_all[valid] / die_area_true[valid])
                    abs_err_all.append(abs_err)
                    log_err_all.append(log_err)
                    ax_abs.scatter(x_all[valid], abs_err, color=scatter_color)
                    ax_log.scatter(x_all[valid], log_err, color=scatter_color)

            elif cfg["type"] == "groupwise_constant":

                for g in group_order:
                    if g not in cfg["groups"]:
                        continue

                    r = cfg["groups"][g]["r"]
                    mask = (P_type_all == g)
                    x = x_all[mask]
                    y_true = die_area_true[mask]
                    y_pred = r * x

                    ax_pred.plot(x_line, r * x_line, "--", color=line_color)
                    ax_pred.scatter(
                        x, y_true,
                        marker=group_markers[g],
                        facecolors="none",
                        edgecolors=edge_color
                    )

                    valid = (y_true > 0) & (y_pred > 0)
                    if valid.any():
                        abs_err = y_pred[valid] - y_true[valid]
                        log_err = np.log10(y_pred[valid] / y_true[valid])
                        abs_err_all.append(abs_err)
                        log_err_all.append(log_err)
                        ax_abs.scatter(x[valid], abs_err,
                                       marker=group_markers[g], color=scatter_color)
                        ax_log.scatter(x[valid], log_err,
                                       marker=group_markers[g], color=scatter_color)

            ax_pred.set_xlabel(cfg["input_label"], fontsize=LABEL_SIZE)

        # =================================================================
        # OUR MODELS (LOG–LOG, PACKAGE AREA)
        # =================================================================
        else:

            x_all = df_test["P_area"].values.astype(float)
            x_line = np.linspace(np.nanmin(x_all), np.nanmax(x_all), 500)

            for g in group_order:
                mask = (P_type_all == g)
                x = x_all[mask]
                y_true = die_area_true[mask]

                if cfg["type"] == "global_log":
                    C, b = 10 ** cfg["a"], cfg["b"]
                elif cfg["type"] == "offset_log":
                    C, b = 10 ** (cfg["a"] + cfg["offsets"][g]), cfg["b"]
                elif cfg["type"] == "groupwise_log":
                    C = 10 ** cfg["groups"][g]["a"]
                    b = cfg["groups"][g]["b"]

                y_pred = C * x ** b

                ax_pred.plot(x_line, C * x_line ** b, color="C0")
                ax_pred.scatter(
                    x, y_true,
                    marker=group_markers[g],
                    facecolors="none",
                    edgecolors="C0"
                )

                valid = (y_true > 0) & (y_pred > 0)
                if valid.any():
                    abs_err = y_pred[valid] - y_true[valid]
                    log_err = np.log10(y_pred[valid] / y_true[valid])
                    abs_err_all.append(abs_err)
                    log_err_all.append(log_err)
                    ax_abs.scatter(x[valid], abs_err,
                                   marker=group_markers[g], color="C0")
                    ax_log.scatter(x[valid], log_err,
                                   marker=group_markers[g], color="C0")

            ax_pred.set_xlabel("Package area [mm²]", fontsize=LABEL_SIZE)

        # =================================================================
        # RMSE + AXES
        # =================================================================
        abs_err_all = np.hstack(abs_err_all) if abs_err_all else np.array([])
        log_err_all = np.hstack(log_err_all) if log_err_all else np.array([])

        rmse_abs = np.sqrt(np.mean(abs_err_all ** 2)) if abs_err_all.size else np.nan
        rmse_log = np.sqrt(np.mean(log_err_all ** 2)) if log_err_all.size else np.nan

        ax_pred.set_title(name, fontsize=TITLE_SIZE)
        ax_abs.set_title(f"RMSE (residual) = {rmse_abs:.3f}", fontsize=12)
        ax_log.set_title(f"RMSE (log) = {rmse_log:.3f}", fontsize=12)

        if loglog:
            ax_pred.set_xscale("log")
            ax_pred.set_yscale("log")
            ax_abs.set_xscale("log")
            ax_log.set_xscale("log")

        all_abs_err.append(abs_err_all)
        all_log_err.append(log_err_all)

    # =====================================================================
    # GLOBAL LIMITS
    # =====================================================================
    abs_all = np.hstack([a for a in all_abs_err if a.size])
    log_all = np.hstack([a for a in all_log_err if a.size])

    abs_max = 1.1 * np.max(np.abs(abs_all))
    log_max = 1.1 * np.max(np.abs(log_all))

    for j in range(n_cols):
        axes[1, j].set_ylim(-abs_max, abs_max)
        axes[2, j].set_ylim(-log_max, log_max)
        axes[1, j].axhline(0, linestyle="--")
        axes[2, j].axhline(0, linestyle="--")
        for i in range(3):
            axes[i, j].grid(True)

    axes[0, 0].set_ylabel("Die area [mm²]")
    axes[1, 0].set_ylabel("Residual (pred − true) [mm²]")
    axes[2, 0].set_ylabel("log10(pred / true)")

    plt.tight_layout()
    plt.show()
```

```python
lit_models_constants = {
    k: literature_models[k]
    for k in [
        "Teehan 2011 (mass)",
        "Teehan 2011 (area)",
        "Our data alpha (LS, area)",
        "Our data alpha (AVG, area)",
        "Proske model",
        "Our data group-wise (AVG, area)",  
    ]
    if k in literature_models
}

plot_groupwise_models(
    our_models={},
    literature_models=lit_models_constants,
    df_test=df_test,
    die_area_true=die_area_true,
    P_type_all=P_type_all,
    loglog=True
)
```

```python
our_models_only = {
    k: our_models[k]
    for k in [
        "Global linear",
        "Linear + category offsets",
        "Group-wise linear (ours)"
    ]
}

plot_groupwise_models(
    our_models=our_models_only,
    literature_models={},
    df_test=df_test,
    die_area_true=die_area_true,
    P_type_all=P_type_all,
    loglog=True
)
```
